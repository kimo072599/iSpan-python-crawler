```python
# 引入 requests 模組
import requests

# 測試用 API 網址
url = 'https://httpbin.org/'

# 發送 GET 請求並取得回應
response = requests.get(url)

print(response)
# 回應的狀態碼
print(response.status_code)

# 回應的原始內容（位元組）
print(response.content)

# 回應的文字內容
print(response.text)
# 會回傳網頁的 html

print(response.headers)
# 以字典的形式回傳
```

```python
# 使用 GET 搭配參數
# https://httpbin.org/get
r = requests.get(url+"get", params={"city": "Taipei"})
# 等同於使用下面的寫法
r2 = requests.get(url+"get"+"?city=Taipei")
print(type(r)) # <class 'requests.models.Response'> 實際上是一個 object
print(r) # <Response [200]>

r_json = r.json()
print(type(r_json),r_json) # 回傳資料是字典(或 list)
print(r_json["args"]) # 可以操作字典，方便爬蟲
r_text = r.text
print(type(r_text), r_text) # 回傳字串
print(r_text["args"]) # 不能操作字串，會報錯

```

```python
# 使用 POST 搭配 data (常用在表單提交)
# https://httpbin.org/post
r = requests.post(url+"post", data={"username": "Kimo"})
print(r.json())
```

```python
# 使用 POST 搭配 JSON (常用在 API 資料傳輸)
r = requests.post(url+"post", json={"username": "Kimo"})
print(r.json())
```
