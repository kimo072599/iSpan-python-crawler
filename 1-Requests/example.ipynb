```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://www.ptt.cc/bbs/Gossiping/index.html"
cookies = {
    "over18" : "1"
}
r = requests.get(url, cookies = cookies)
soup = bs(r.text, "lxml")

print(soup)
# 到這邊是確定有爬到網頁 html，接下來觀察 html 還是要去網頁看比較好看
```

```python
containers = soup.select(".r-ent")
# print(containers)
# print(containers[0])
containers = soup.select(".r-ent .title")
# print(containers)
# print(containers[0])
title_elements = soup.select(".r-ent .title a")
print(title_elements)

# 因為 href 裡沒有給完整網址，所以要自己給前贅
prefix = "https://www.ptt.cc"

for ele in title_elements:
    print(f"標題：{ele.get_text()} 網址：{prefix + ele["href"]}")

    ppt_response = requests.get(prefix + ele["href"])
    print(ppt_response.)
```
