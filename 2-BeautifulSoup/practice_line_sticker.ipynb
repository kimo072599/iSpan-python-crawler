```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://store.line.me/stickershop/product/3807122/zh-Hant?from=sticker"

response = requests.get(url)
page_html = bs(response.text, "lxml")
print(page_html)
```

```python
# 找到 stickers 的 element
sticker_list = page_html.find("ul", class_="mdCMN09Ul FnStickerList")
sticker_elements = sticker_list.find_all("li")
print(len(sticker_elements))
print(sticker_elements[0])
```

```python
# 建立資料夾
import os
folder_name = "line_stickers"

print(os.path.exists(folder_name))

if not os.path.exists(folder_name):
    os.mkdir("line_stickers")
else:
    print(f"以建立過{folder_name}")
```

```python
# 自動建立 folder_name
sticker_title_element = page_html.find("p", {"data-test":"sticker-name-title"})
folder_name = sticker_title_element.text
print(folder_name)
```

```python
# 取得所有 stickers 的路徑
import json

for index, sticker_element in enumerate(sticker_elements, start=1):
    # print(sticker_element)
    # print(sticker_element["data-preview"])
    # data_preview = sticker_element["data-preview"]
    # 這樣去取用會報錯，要轉 json 格式
    data_preview = json.loads(sticker_element["data-preview"])
    # print(data_preview)
    static_url = data_preview["staticUrl"]
    print(static_url)

    # 將圖片寫入檔案
    image_content = requests.get(static_url).content
    with open(f"line_stickers/line_sticker_{index}.png", "wb") as f:
        f.write(image_content)
```
