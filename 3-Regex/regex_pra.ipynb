```python
import re
text = " ID/姓名： 123456789 / Denji"

# 使用正則表達式提取 ID 和姓名
regex = r"ID/姓名： (\d+) / (\w+)"

# 使用 search / match 方法進行匹配
match = re.match(regex, text) # 從字串開頭開始比對
search = re.search(regex, text)

print(match)
print(search)
print(search.group(0))
print(search.group(1))
print(search.group(2))
```

```python
import re
text = " ID/姓名： 123456789 / Denji"

# 使用正則表達式提取 ID 和姓名
regex = r"ID/姓名： (\d+) / (\w+)"

data = re.findall(regex, text)
print(data) # 回傳一個 tuple
for i in data:
    print(f"ID: {i[0]}")
    print(f"name: {i[1]}")
for id, name in data:
    print(f"ID: {id}")
    print(f"name: {name}")
```

### ⚠️ 要使用 `.`、`(`、`)` 時要使用 `\` 跳脫

```python
import re

text = '住家電話：(02)2311-3838'
regex = r"\((\d+)\)(\d+-\d+)"
# 使用正則表達式提取區碼和電話號碼
data = re.findall(regex, text)
for i in data:
    print(f"區碼: {i[0]}")
    print(f"電話號碼: {i[1]}")
```

```python
text = """
    請用以下方式聯絡：
    住家: (02)2311-3838
    公司: (02)2311-3839#345
    手機: 0912-345-678
"""

# 使用正則表達式提取多種電話號碼格式
regex = r"(\(\d+\))?(\d+-\d+(-\d+)?)(#\d+)?"
# 這樣會很多群組，在印的時候會很麻煩
data = re.findall(regex, text)
print(data)
for i in data:
    print(f"{i[0]}{i[1]}{i[3]}")

# 使用?:讓群組不顯示
regex = r"(?:\(\d+\))?\d+-\d+(?:-\d+)?(?:#\d+)?"

data = re.findall(regex, text)
print(data)
for i in data:
    print(i)

# 酷方法
regex = r"\(?\d+.\d+.\d+.\d+"
data = re.findall(regex, text)
print(data)
```

```python
for match in matches[2:]:
    pattern_td = r"<td>(.*?)</td>"
    info_list = re.findall(pattern_td, match)
    print(info_list)
```

```python
grade_dict ={
    0: "S",
    1: "A",
    2: "B",
    3: "C",
    4: "D"
}
anime_grade = {}
for match in matches[2:]:
    pattern_td = r"<td>(.*?)</td>"
    info_list = re.findall(pattern_td, match)
    for index, value in enumerate(info_list[2:]):
        if value == "★":
            # print(grade_dict[index])
            anime_grade[info_list[1]] = grade_dict[index]
print(anime_grade)
```
