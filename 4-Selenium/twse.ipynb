```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import os
import json

# 設定 Chrome 瀏覽器的選項
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized") # Chrome 瀏覽器在啟動時最大化視窗
options.add_argument("--incognito") # 無痕模式
options.add_argument("--disable-popup-blocking") # 停用 Chrome 的彈窗阻擋功能。

# 建立 Chrome 瀏覽器物件
driver = webdriver.Chrome(options=options)
driver.get("https://www.twse.com.tw/zh/trading/historical/stock-day-avg.html")
```

```python
def get_stock_price(code):
    driver.find_element(By.CSS_SELECTOR, ".stock-code-autocomplete").send_keys(code)
    driver.find_element(By.CSS_SELECTOR, ".submit button").click()
    elements = driver.find_elements(By.CSS_SELECTOR, "tbody.is-last-page > tr")

    for ele in elements[:-1]:
        print(ele.text)

get_stock_price("0050")
```
